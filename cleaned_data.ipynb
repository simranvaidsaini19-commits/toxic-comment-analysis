```
import pandas as pd
import numpy as np
```

```
df=pd.read_csv('/content/sample_data/archive.zip')
```

```
df=df.dropna()
```

```
df=df.drop_duplicates()
```

```
df['comment_text']=df['comment_text'].astype(str).str.lower()
```

```
import re
df['comment_text']=df['comment_text'].apply(lambda x:re.sub(r'[^a-zA-Z]','',x))
```

```
df['comment_text']=df['comment_text'].str.strip()
```

```
df.info()
```

```
df.to_csv('cleaned_dataset.csv',index=False)
```
